```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
spark = SparkSession.builder.appName('json_transformation').getOrCreate()
```

```python
sjson = spark.read.format('json').option('multiline',True).load('/Volumes/workspace/default/spark_data/simple_json.json')
sjson.display()
```

```python
sjson.select('address.city','address.state','address.street','customer_id','name','phone').display()
```

```python
sjson2 = spark.read.format('json').option('multiline',True).load('/Volumes/workspace/default/spark_data/simp_json2.json')
sjson2.display()
```

```python
sjson3 = sjson2.select(col('customer.customerid'),col('customer.name').alias('cname'),'customer.address.city','customer.address.state','order.orderdate','order.totalAmount',explode(col('order.items.productName')).alias('productName'),explode(col('order.items.quantity')).alias('quantity'),explode(col('order.items.price')).alias('price') )
#explode(col('order.items.supplier.name')).alias('spl_name'))
sjson3.display()
```

```python
sjson4 = sjson3.withColumn('productName',explode(col('productName')) ).withColumn('quantity',explode(col('quantity')))
sjson4.display()
```

```python
sjson4.dropDuplicates().display()
```
